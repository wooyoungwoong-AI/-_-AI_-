### 목적
프로젝트의 주요 목적 중 하나는 시각 장애인이 일상 생활에서 더욱 독립적으로 활동할 수 있도록 돕는 것입니다. 시각 장애인은 이미지와 텍스트 정보에 대한 접근에 제한이 있을 수 있으므로 이를 보완하여 필요한 정보를 쉽게 얻을 수 있도록 합니다.

### 기술 활용 방식
ChatGPT API와 음성인식, 번역, TTS 기능을 활용하여 prompt를 직접 입력할 필요 없이 음성으로만 ChatGPT의 서비스를 이용할 수 있도록 합니다. 또한 촬영한 이미지(이미지 인식 기능 예선에선 미구현 본선 구현 예정) 또는 prompt의 입력으로 사용하기 어렵거나 위험한 전자제품 등의 기구에 대한 사용 방법 및 주의할 점을 안내합니다.

1. 시각 장애인 들은 gpt처럼 텍스트로 된 정보를 이해하기 힘듭니다. 이런 문제점을 해결할 방안으로 음성인식을 사용해 음성을 gpt의 input으로 받고 텍스트 형태인 output 값을 TTS로 설명해 줍니다. 뉴스, 문서등 다양한 정보를 보다 쉽게 얻을 수 있어 정보 격차 문제를 해결할 수 있습니다.
1. 앱 기반 TTS AI는 가전 제품을 볼 수 없어 사용법을 이해하기 힘든 시각 장애인들을 위한 prompt를 작성하고 이를 기반으로 시각 장애인들이 이해하기 힘들고 사고가 발생할 수 있는 위험한 가전 제품의 주의 사항과 사용법을 시각장애인에 눈높이에 맞춰 설명해 줄 수 있습니다.
1. 시각 장애인 이라는 제약 사항으로 인해 몇몇 직업에 국한되어 있는 현재의 문제점을 보완해 여러 분야에 도전할 수 있도록 도와줍니다.
1. 음성 인식을 통해 글을 작성해 주고 최신 정보를 생성해 주기 때문에 인터넷에서 다른 사람들과 교류하기 쉽게 진입 장벽을 낮추어 사회에 더 활발하게 참여할 수 있게 도와줍니다.
1. 인터넷에서 제공하는 정보들은 시각 장애인의 입장에서 이해하기 힘든 내용이 있을 수 있습니다. 이런 문제를 해결하기 위해 시각 장애인의 눈높이에 맞춰 이해하기 쉽게 설명해 시각 장애인들이 자립성을 키울 수 있도록 돕습니다.

```python
r = sr.Recognizer()
```

```python
# 음성 인식

def Recording(audio_file):
  try:
    harvard = sr.AudioFile(audio_file)
    with harvard as source:
      audio = r.record(source)
    return r.recognize_google(audio, language='ko_KR')
  except:
    speak('''음성 파일을 불러드리는데 오류가 발생하였습니다.
    음성 파일의 길이나 잡음의 문제일 수 있으니 다시 녹음해 주세요.''')
```

Translator Part


```python
# papago API
# 번역 기능으로 papago가 한국어를 다른 언어로 번역하는 능력이 뛰어나다 해서 번역 기능으로 papago를 채택하게 되었습니다.

# translate 함수 선언
def translator(text):
  client_id = ""
  client_secret = ""
  data = {'text' : text,
          'source' : 'ko',
          'target' : 'en'}
  url = "https://openapi.naver.com/v1/papago/n2mt"
  header = {"X-Naver-Client-id":client_id,
            "X-Naver-client-Secret":client_secret}
  response = requests.post(url, headers=header, data=data)
  rescode = response.status_code
  if(rescode==200):
    send_data = response.json()
    trans_data = (send_data['message']['result']['translatedText'])
    return trans_data
  else:
    print("Error Code:", rescode)
```
